```python
country_plot=df.pivot_table(index='COUNTRY', values='SALES', aggfunc=np.sum)
country_plot.sort_values('SALES',inplace=True,ascending=False)
country_plot.plot(kind='bar',figsize=(13,7))
plt.xlabel('COUNTRY')
plt.ylabel('SALES')
plt.title('Country and Sales Analysis')
plt.xticks(rotation=80)
plt.show()
```

```python
country_plot=df.pivot_table(index='MONTH_ID', values='SALES', aggfunc=np.sum)
country_plot.sort_values('SALES',inplace=True,ascending=False)
country_plot.plot(kind='bar',figsize=(13,7))
plt.xlabel('MONTH')
plt.ylabel('SALES')
plt.title('MONTH and Sales Analysis')
plt.xticks(rotation=80)
plt.show()
```
